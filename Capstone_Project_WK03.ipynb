Using Beautiful Soup library for scrapping of table from wikipedia webpage. Then title of the webspage is printed to see if the scrapping has been done successfully or not. Finally the table is printed.

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
table = str(soup.table)
display_html(table,raw=True)
```

This table is converted into pandas dataframe for wranggling and pre-processing

```python
df_new = pd.read_html(table)
df=df_new[0]
df.head()
```

```python
# Only process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned
df_temp = df[df.Borough != 'Not assigned']

# Combining the neighborhoods with same Postal Code
df_final = df_temp.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df_final.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
df_final['Neighborhood'] = np.where(df_final['Neighborhood'] == 'Not assigned',df_final['Borough'], df_final['Neighborhood'])

df_final.head()
```

```python
#shape of the final datagframe after cleaning

df_final.shape
```

Importing the .csv file containing coordinates (latitude and longitude) information for neighborhoods in Canada

```python
location = pd.read_csv('https://cocl.us/Geospatial_data')
location.head()
```

Merging the two dataframes df_final and location to get the coordinates of neighborhoods in Canada

```python
df_loc = pd.merge(df_final, location, on = 'Postal Code')
df_loc.head()
```

Getting all the rows from this dataframe which contains Toronto in its Borough

```python
df_Toronto = df_loc[df_loc['Borough'].str.contains('Toronto',regex=False)]
df_Toronto
```

let's get the top 10 venues that are in Toronto within a radius of 500 meters.
First, let's create the GET request URL. Name your URL url.

```python
LIMIT = 100 # limit of number of venues returned by Foursquare API

radius = 500 # define radius

# create URL
url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
    CLIENT_ID, 
    CLIENT_SECRET, 
    VERSION, 
    latitude, 
    longitude, 
    radius, 
    LIMIT)
url
```

Send the GET request to examine the results

```python
results = requests.get(url).json()
results
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

Now clean the json and structure it into a pandas dataframe.

```python
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```
